###$\text{1.5. Model Regresi}$

####$\text{a.) Regresi C terhadap Y}$

```python
#Import library
from sklearn.linear_model import LinearRegression

x = data["Y"].values        #data feature
y = data["C"].values        #data label

x = x.reshape(-1, 1)      #1D --> 2D

#Membuat objek model
regressor = LinearRegression()
regressor.fit(x, y)       #Model fitting

#Plot Data
fig, ax = plt.subplots(figsize=(16, 8))
mapbar_1 = ax.scatter(data["Y"], data["C"], s = 10**(-9.8)*data["C"], c=data["C"], cmap = "viridis")
ax.plot(data["Y"], regressor.predict(x), color="red")
ax.set_title("C terhadap GDP (Y)")
ax.set_xlabel("GDP (Y)")
ax.set_ylabel("C")
fig.colorbar(mapbar_1)
plt.show()

#Parameter Regresi
print("Slope:", regressor.coef_)
print("Intercept:", regressor.intercept_, "\n", "\n")
```

Jadi, persamaan regresinya 
$$C = 2244338675.3159 + 0.5622Y$$

####$\text{b.) Regresi I terhadap Y dan r}$

```python
#Import library
from sklearn.linear_model import LinearRegression

x = data[["Y", "r"]]        #data feature
y = data["I"].values        #data label

#Membuat objek model
regressor = LinearRegression()
regressor.fit(x, y)       #Model fitting

#Plot Data
fig, ax = plt.subplots(figsize=(16, 8))
mapbar_2 = ax.scatter(data["Y"], data["I"], s = 20*data["r"], c = data["r"], cmap = "winter")
                                                       #Nilai dari r direpresentasikan dengan ukuran 
                                                       #dan warna dari scatter plot
ax.plot(data["Y"], regressor.predict(x), color="red")
ax.set_title("I terhadap GDP (Y) dan Suku Bunga (r)")
ax.set_xlabel("GDP (Y) & Suku Bunga (r)")
ax.set_ylabel("I")
fig.colorbar(mapbar_2)
plt.show()

#Parameter Regresi
print("Slope:", regressor.coef_)
print("Intercept:", regressor.intercept_, "\n", "\n")
```

Jadi, persamaan regresinya 
$$I = -4120368377.6461 + 0.2585Y + (7.4777 \text{ x } 10^{8})r$$

####$\text{c.) Regresi Y terhadap G dan M}$

```python
#Import library
from sklearn.linear_model import LinearRegression

x = data[["G", "M"]]        #data feature
y = data["Y"].values        #data label

#Membuat objek model
regressor = LinearRegression()
regressor.fit(x, y)       #Model fitting
y_predict = regressor.predict(x)

#Plot Data
fig, ax = plt.subplots(figsize=(16, 8))
mapbar_3 = ax.scatter(data["G"], data["Y"], s = 12**(-9)*data["M"], c = data["M"], cmap = "inferno")
                                                 #Nilai dari M direpresentasikan dengan ukuran 
                                                 #dan warna dari scatter plot
ax.plot(data["G"], y_predict, color="red")
ax.set_title("Y terhadap Belanja Pemerintah (G) dan Jumlah Uang Beredar (M)")
ax.set_xlabel("Belanja Pemerintah (G) dan Jumlah Uang Beredar (M)")
ax.set_ylabel("Y")
fig.colorbar(mapbar_3)
plt.show()

#Parameter Regresi
print("Slope:", regressor.coef_)
print("Intercept:", regressor.intercept_, "\n", "\n")
```

Jadi, persamaan regresinya 
$$Y = 1653884933.0614 + 5.8943G - 0.0689M$$

####$\text{Prediksi efek multiplier}$

Dengan asumsi ada keterlibatan pihak asing (***Ekonomi Terbuka***): $$$$

$$Y = C + I + G + NX$$ $$$$

$$Y = (c_0 + c_1Y) + (i_0 + i_1Y + i_2r) + G + (X - IM)$$ $$$$

$$Y(1 - c_1 - i_1) = c_0 + i_0 + i_2r + G + (X - IM)$$ $$$$

$$Y = \frac{1}{(1 - c_1 - i_1)}[c_0 + i_0 + i_2r + G + (X - IM)]$$ $$$$

Jadi, multipliernya adalah 
$$\frac{1}{1 - c_1 - i_1} = \frac{1}{1 - (0.56229454) - (0.259291961)} = 5.6049$$

#$\text{II. Model AD-AS}$

$$\begin{cases}
AD: Y &= \frac{c_0 - c_1 + c + \bar{G} + \frac{b}{e} \left(\frac{M^d}{P} - f \right)}{\left(1 - c_1 - a - \frac{bd}{e} \right)} \\
AS: P &= P^e(1 + \mu)\left(1 - \alpha \left[1 - \frac{Y}{L} \right] + z\right]
\end{cases}$$
